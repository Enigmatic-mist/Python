```python
#Importing various libraries
import numpy as np
import matplotlib.pyplot as plt
%matplotlib ipympl
from scipy.optimize import curve_fit
```

```python
#Defining a function to calculate the specral radiance per unit wavelength
def srpuf(lam,h,c,kb,t):
    return 2*h*c*c/(((np.exp(h*c/(lam*kb*t)) - 1))*(lam**5))
# A function to read the contents in a file and put them in their appropriate arrays
def read_file(fname) :
    z = []
    lam =[]
    with open(fname) as file:
        for line in file:
            lam.append(float(line.replace(","," ").split()[0]))
            z.append(float(line.replace(","," ").split()[1]))
    return np.array(lam),np.array(z)
# Reading the first file
lam1,z1 = read_file("d1.txt")
# Assuming some estimate values
est_values = [6.626e-34, 3.0e8, 1.38e-23, 300] 
# Defining a function to calculate the amount of error
def error(est,value):
    return ((est-value)/est)*100
```
